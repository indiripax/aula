### Neste caderno iremos:
1. *Importar* o pacote pandas com o nome **pd**
2. *Criar* uma tabela de dados (**dataframe**) nova 3x3 com dados exemplo 
3. *Nomear* **colunas** e **linhas** de uma tabela de dados.
4. *Adicionar* ou *Remover* **colunas** ou **linhas** de uma tabela de dados
5. *Carregar* ou *Salvar* uma tabela de dados em formato ***csv***

```python
# Importando o pacote pandas
import pandas as pd
```

```python
# Criando uma tabela de dados 3x3 diretamente
tabela_1 = pd.DataFrame([['maça','amarelo',23],['goiaba','verde',11], ['banana', 'azul', 9]])
```

```python
# Visualizamos a tabela criada. Observe a ausência de nomes nas linhas e colunas
tabela_1
```

```python
# Para melhorar nossa orientação, damos nomes às linhas e colunas
tabela_1.columns = ['Fruta', 'Cor', 'Dias']
tabela_1.index = ['P-01', 'P-02', 'P-03']
```

```python
# Visualizamos novamente os dados...
tabela_1
```

```python
# Adicionar nova coluna de dados
tabela_1['Valor'] = [22.50, 1.99, 2.99]
```

```python
# Adicionar nova linha de dados
tabela_1.loc['P-04'] = ['laranja', 'violeta', '2', '7.31']
```

```python
# Visualizamos novamente os dados...
tabela_1
```

```python
# Remover uma coluna de dados
tabela_2 = tabela_1.drop('Cor', axis='columns')
tabela_2
```

```python
# Remover uma linha de dados
tabela_3 = tabela_1.drop('P-02', axis='index')
tabela_3
```

```python
# Salvando uma tabela de dados em formato csv
tabela_1.to_csv('tabela_1.csv')
```

```python
# Carregando uma tabela de dados em formato csv para uma tabela pandas
# Atenção para os caracteres separadores de colunas e decimal!!!
iris_data = pd.read_csv("iris_dataset.csv", sep=';', decimal=',', encoding='latin-1', index_col=0)
iris_data
```
